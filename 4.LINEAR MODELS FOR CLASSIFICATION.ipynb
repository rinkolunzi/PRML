```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
data,target=datasets.make_classification(n_samples=100, 
                                         n_features=2,
                                         n_informative=2,
                                         n_redundant=0,
                                         n_repeated=0, 
                                         n_classes=2,
                                         n_clusters_per_class=1)

plt.scatter(data[:,0],data[:,1],c=target)
plt.show()
```

```python
clf = LinearRegression()
clf.fit(data,target)
x, y = np.mgrid[-4:4:.02, -4:4:.02]
x_test = np.array([x,y]).reshape(2, -1).T
z=clf.predict(x_test)
z=np.array(z).reshape(400,400)
plt.contour(x,y,z,levels=np.linspace(0.2*z.max(), z.max(), 1))

plt.scatter(data[:,0],data[:,1],c=target)
plt.show()
```

```python
data,target=datasets.make_classification(n_samples=100, 
                                         n_features=2,
                                         n_informative=2,
                                         n_redundant=0,
                                         n_repeated=0, 
                                         n_classes=2,
                                         n_clusters_per_class=1)
plt.scatter(data[:,0],data[:,1],c=target)
plt.show()
```

```python
##linear regression
def fit(X,t):
    pse=np.linalg.pinv(X)#Compute the Moore-Penrose pseudo-inverse
    w=pse@t
    return w
def predict(X):
    w=fit(data,target)
    return X@w
x, y = np.mgrid[-4:4:.02, -4:4:.02]
x_test = np.array([x,y]).reshape(2, -1).T
z=predict(x_test)
a=[]
for item in z:
    if item>=0:
        a.append(1)
    else:
        a.append(0)
a=np.array(a).reshape(400,400)
plt.scatter(data[:,0],data[:,1],c=target)
plt.contour(x,y,a)
plt.show()
```

```python
#Logistic regression
#coding without package
#(4.96 4.97)
data,target=datasets.make_classification(n_samples=100, 
                                         n_features=2,
                                         n_informative=2,
                                         n_redundant=0,
                                         n_repeated=0, 
                                         n_classes=2,
                                         n_clusters_per_class=1)

plt.scatter(data[:,0],data[:,1],c=target)
plt.show()
```

```python
def sigma(x):
    return 1/(1+np.exp(-x))
def fit(X,t):
    w = np.zeros(np.size(X, 1))
    for i in range(100):
        y = sigma(X @ w)
        grad = X.T @ (y - t)
        hessian = (X.T * y * (1 - y)) @ X
        w=w-np.linalg.inv(hessian)@grad
        #w=w -np.linalg.solve(hessian, grad)
    return w
def predict(X):
    w=fit(data,target)
    y=sigma(X@w)
    threshold=0.5
    label = (y > threshold).astype(np.int)
    return label
x, y = np.mgrid[-4:4:.02, -4:4:.02]
x_test = np.array([x,y]).reshape(2, -1).T
z=predict(x_test)
a=np.array(z).reshape(400,400)
plt.scatter(data[:,0],data[:,1],c=target)
plt.contour(x,y,a)
plt.show()
```

```python
#probablity logistic regression
def proba(X):
    w=fit(data,target)
    y=sigma(X@w)
    return y
x, y = np.mgrid[-4:4:.02, -4:4:.02]
x_test = np.array([x,y]).reshape(2, -1).T
z=proba(x_test)
a=np.array(z).reshape(400,400)
plt.contourf(x,y,a,levels=np.linspace(0.2*z.min() +0.2*z.max(), 0.1*z.min() +0.4*z.max(), 7),cmap=plt.cm.Blues)
plt.scatter(data[:,0],data[:,1],c=target)
plt.colorbar()
plt.show()
```
